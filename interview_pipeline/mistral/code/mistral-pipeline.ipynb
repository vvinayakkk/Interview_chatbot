````python
import torch
import pandas as pd
import numpy as np
import re
import textwrap
from transformers import AutoModelForCausalLM, AutoTokenizer
from peft import PeftModel
import google.generativeai as genai
from langchain_google_genai import ChatGoogleGenerativeAI
from sentence_transformers import SentenceTransformer
from faiss import IndexFlatL2
import random
import seaborn as sns
import matplotlib.pyplot as plt
from typing import List, Dict, Tuple
import json
from datetime import datetime
import os
import plotly.express as px
import plotly.graph_objects as go
from plotly.subplots import make_subplots
import pandas as pd
from unsloth import FastLanguageModel
import torch

class InterviewPipeline:
    def __init__(
    self,
    model_name: str,  # Replace gemma_adapter_path
    google_api_key: str,
    passage_db_path: str,
    max_seq_length: int = 700,  # Moved after required arguments
    interview_type: str = 'mediocre',
    save_dir: str = '/content/interview_results'
):
        """Initialize the interview pipeline with enhanced tracking and configuration"""
        # Initialize models and components
        self.sentence_model = SentenceTransformer('all-MiniLM-L6-v2')
        self.passages_df = pd.read_csv(passage_db_path)
        self.setup_faiss_index()

        # Load question generation model
        # Replace these lines in __init__
        self.question_model, self.tokenizer = self.load_question_model(
            model_name=model_name,
            max_seq_length=max_seq_length
        )
                # Initialize Gemini
        genai.configure(api_key=google_api_key)
        self.llm = ChatGoogleGenerativeAI(
            model="gemini-pro",
            google_api_key=google_api_key,
            temperature=0.7
        )

        # Interview configuration
        self.interview_type = interview_type
        self.correct_threshold = {
            'good': 0.8,
            'mediocre': 0.6,
            'poor': 0.4
        }
        self.base_correct_prob = {
            'good': 0.9,
            'mediocre': 0.7,
            'poor': 0.4
        }

        # Interview state tracking
        self.current_difficulty = 'medium'
        self.current_topic = None
        self.asked_questions = []
        self.covered_topics = set()
        self.total_score = 0
        self.question_count = 0

        # Enhanced difficulty adjustment tracking
        self.performance_history = []
        self.topic_scores = {}
        self.consecutive_success = 0
        self.consecutive_failure = 0
        self.adjustment_logs = []

        # Session metrics
        self.session_start_time = None
        self.question_times = []
        self.topic_coverage = {}
        self.difficulty_changes = []
        self.performance_log = []

        # Save configuration
        self.save_dir = save_dir
        self.session_id = datetime.now().strftime('%Y%m%d_%H%M%S')
        os.makedirs(os.path.join(save_dir, self.session_id), exist_ok=True)

        # Configuration
        self.config = {
            'min_questions_per_difficulty': 2,
            'max_consecutive_same_topic': 3,
            'difficulty_change_threshold': 0.2,
            'topic_mastery_threshold': 0.8,
            'max_time_per_question': 300
        }

    # [... rest of your existing methods from setup_faiss_index() to _display_session_summary() ...]
    def setup_faiss_index(self):
        """Initialize FAISS index for passage similarity search"""
        passages = self.passages_df['Retrieved_Passage'].tolist()
        embeddings = self.sentence_model.encode(passages)

        self.index = IndexFlatL2(embeddings.shape[1])
        self.index.add(embeddings.astype('float32'))
        self.passage_embeddings = embeddings


    def load_question_model(self, model_name: str, max_seq_length: int =700 ):
        """Load the Mistral model using Unsloth"""
        try:
            print("Loading model and tokenizer...")
            dtype = None  # None for auto detection
            load_in_4bit = True
            
            model, tokenizer = FastLanguageModel.from_pretrained(
                model_name=model_name,
                max_seq_length=max_seq_length,
                dtype=dtype,
                load_in_4bit=load_in_4bit,
            )
            
            FastLanguageModel.for_inference(model)
            print("Model loaded successfully!")
            return model, tokenizer
        except Exception as e:
            print(f"Error loading model: {str(e)}")
            import traceback
            traceback.print_exc()
            return None, None


    def _score_to_difficulty(self, score: float) -> float:
        """Convert a score to a difficulty value between 0 and 2"""
        if score >= 0.8:
            return 2.0  # Suggest hard
        elif score >= 0.6:
            return 1.0  # Suggest medium
        else:
            return 0.0  # Suggest easy

    def _get_consecutive_difficulty(self) -> float:
        """Calculate difficulty based on consecutive performance"""
        if self.consecutive_success >= 2:
            return 2.0  # Suggest hard after 2 consecutive successes
        elif self.consecutive_failure >= 2:
            return 0.0  # Suggest easy after 2 consecutive failures
        return 1.0  # Suggest medium otherwise

    def _get_final_difficulty(self, weighted_score: float) -> str:
        """Convert weighted score to final difficulty with stability rules"""
        # Apply stability rules
        if self.current_difficulty == 'hard' and weighted_score < 1.5:
            # Require significant evidence to drop from hard
            return 'medium' if weighted_score < 1.0 else 'hard'
        elif self.current_difficulty == 'easy' and weighted_score > 0.5:
            # Require significant evidence to rise from easy
            return 'medium' if weighted_score > 1.0 else 'easy'

        # Standard conversion
        if weighted_score >= 1.5:
            return 'hard'
        elif weighted_score >= 0.5:
            return 'medium'
        else:
            return 'easy'

    def _log_difficulty_adjustment(self, difficulty_scores: dict, weighted_score: float,
                                new_difficulty: str) -> None:
        """Log difficulty adjustment details for debugging and analysis"""
        log_data = {
            'timestamp': pd.Timestamp.now(),
            'old_difficulty': self.current_difficulty,
            'new_difficulty': new_difficulty,
            'factor_scores': difficulty_scores,
            'weighted_score': weighted_score,
            'consecutive_success': self.consecutive_success,
            'consecutive_failure': self.consecutive_failure,
            'performance_history': len(self.performance_history)
        }

        # Store log data (implement storage method as needed)
        if not hasattr(self, 'adjustment_logs'):
            self.adjustment_logs = []
        self.adjustment_logs.append(log_data)

    def adjust_difficulty(self, score: float) -> str:
      """
      Advanced difficulty adjustment system that considers multiple factors:
      - Recent performance trend
      - Topic mastery
      - Consecutive performance
      - Time-based factors
      - Overall session performance
      """
      # Add these as instance variables in __init__
      if not hasattr(self, 'performance_history'):
          self.performance_history = []
      if not hasattr(self, 'topic_scores'):
          self.topic_scores = {}
      if not hasattr(self, 'consecutive_success'):
          self.consecutive_success = 0
      if not hasattr(self, 'consecutive_failure'):
          self.consecutive_failure = 0

      # Update performance history
      self.performance_history.append({
          'score': score,
          'difficulty': self.current_difficulty,
          'topic': self.current_topic if hasattr(self, 'current_topic') else None
      })

      # 1. Calculate recent performance trend (last 3 questions)
      recent_scores = [x['score'] for x in self.performance_history[-3:]]
      trend = np.mean(recent_scores) if len(recent_scores) >= 3 else score

      # 2. Update topic mastery
      current_topic = self.current_topic if hasattr(self, 'current_topic') else 'default'
      if current_topic not in self.topic_scores:
          self.topic_scores[current_topic] = []
      self.topic_scores[current_topic].append(score)

      # Calculate topic mastery level
      topic_mastery = np.mean(self.topic_scores[current_topic])

      # 3. Update consecutive performance
      if score >= 0.8:
          self.consecutive_success += 1
          self.consecutive_failure = 0
      elif score <= 0.4:
          self.consecutive_failure += 1
          self.consecutive_success = 0
      else:
          self.consecutive_success = 0
          self.consecutive_failure = 0

      # 4. Calculate overall session performance
      overall_performance = np.mean([x['score'] for x in self.performance_history])

      # 5. Calculate difficulty adjustment factors
      difficulty_scores = {
          'trend': self._score_to_difficulty(trend),
          'topic_mastery': self._score_to_difficulty(topic_mastery),
          'consecutive': self._get_consecutive_difficulty(),
          'overall': self._score_to_difficulty(overall_performance)
      }

      # 6. Apply weights to different factors
      weights = {
          'trend': 0.35,
          'topic_mastery': 0.25,
          'consecutive': 0.25,
          'overall': 0.15
      }

      # 7. Calculate weighted difficulty score
      weighted_score = sum(difficulty_scores[factor] * weight
                          for factor, weight in weights.items())

      # 8. Apply stability rules
      new_difficulty = self._get_final_difficulty(weighted_score)

      # 9. Log adjustment details if needed
      self._log_difficulty_adjustment(difficulty_scores, weighted_score, new_difficulty)

      return new_difficulty


    def select_next_topic(self) -> Tuple[str, str]:
        """Select next topic and subtopic ensuring coverage"""
        available_topics = set(zip(
            self.passages_df['Topic'],
            self.passages_df['Sub-Topic']
        )) - self.covered_topics

        if not available_topics:
            self.covered_topics.clear()
            available_topics = set(zip(
                self.passages_df['Topic'],
                self.passages_df['Sub-Topic']
            ))

        return random.choice(list(available_topics))

    def find_similar_passages(self, query: str, k: int = 3) -> List[Dict]:
        """Find similar passages using FAISS"""
        query_embedding = self.sentence_model.encode([query])
        D, I = self.index.search(query_embedding.astype('float32'), k)

        similar_passages = []
        for idx in I[0]:
            passage_data = self.passages_df.iloc[idx]
            similar_passages.append({
                'passage': passage_data['Retrieved_Passage'],
                'topic': passage_data['Topic'],
                'subtopic': passage_data['Sub-Topic'],
                'difficulty': passage_data['Difficulty']
            })

        return similar_passages

    def generate_question(self, passage: str, topic: str, subtopic: str, difficulty: str) -> str:
        """Generate a question using the Mistral model"""
        prompt = f"""Generate a concise single question based on the following passage:
    Topic: {topic}
    Subtopic: {subtopic}
    Difficulty: {difficulty}
    Passage: {passage}
    Question:"""
    
        inputs = self.tokenizer(prompt, return_tensors="pt").to(self.question_model.device)
        
        with torch.no_grad():
            outputs = self.question_model.generate(
                inputs["input_ids"],
                max_length=700,
                temperature=0.7,
                do_sample=True,
                top_p=0.9,
                top_k=50
            )
        
        question = self.tokenizer.decode(outputs[0], skip_special_tokens=True)
        question = question[question.find("Question:") + len("Question:"):].strip()
        
        return question if question.endswith('?') else question + '?'
        
    def display_question_result(self, result: Dict):
      """Format and display the result of a single question"""
      print("\n" + "="*80)
      print(f"Topic: {result['topic']} - {result['subtopic']}")
      print(f"Difficulty: {result['difficulty']}")
      print("-"*80)

      print("Question:")
      print(result['question'])
      print("\nGiven Answer:")
      print(result['answer'])
      print("\nCorrect Answer:")
      print(result['correct_answer'])

      print("\nEvaluation:")
      print(f"Score: {result['score']:.2f}/1.00")
      print("\nFeedback:")
      print(textwrap.fill(result['feedback'], width=80))

      print("="*80)

    def clean_response_for_json(self, text: str) -> str:
      """Clean the response text to make it JSON-compatible"""
      # Remove markdown code blocks
      text = re.sub(r'```[^`]*```', '', text)
      # Remove single backticks
      text = text.replace('`', '')
      # Replace newlines with spaces
      text = text.replace('\n', ' ')
      # Replace multiple spaces with single space
      text = re.sub(r'\s+', ' ', text)
      # Escape quotes
      text = text.replace('"', '\\"')
      return text

    def generate_answer(self, question: str, passage: str, should_be_correct: bool) -> str:
        """Generate an answer using Gemini with enforced score ranges"""
        # Define correct answer probabilities for each interview type
        correct_probabilities = {
            'good': 0.7,    # 70% correct, 30% wrong
            'mediocre': 0.5,  # 50% correct, 50% wrong
            'poor': 0.1     # 10% correct, 90% wrong
        }
    
        # Determine if this answer should be correct based on interview type
        should_be_correct = random.random() < correct_probabilities[self.interview_type]
    
        if should_be_correct:
            if self.interview_type == 'good':
                score_range = "between 0.8 and 1.0"
            elif self.interview_type == 'mediocre':
                score_range = "between 0.6 and 0.8"
            else:  # poor
                score_range = "between 0.4 and 0.6"
                
            prompt = f"""Based on this reference passage: "{passage}"
    
    Generate an answer to this question: "{question}" that would score {score_range} out of 1.0.
    
    Requirements:
    1. Include enough correct information to score within the specified range
    2. For non-perfect scores, include minor errors or omissions
    3. Match the quality level to the score range
    4. If code examples are needed, use proper syntax
    
    Generate your answer now:"""
        else:
            if self.interview_type == 'poor':
                score_range = "between 0.0 and 0.2"
                prompt = f"""Generate a completely incorrect answer to this question: "{question}" that would score {score_range} out of 1.0.
    
    Requirements:
    1. Answer should be mostly or entirely wrong
    2. Use incorrect terminology and concepts
    3. Make statements that contradict the actual topic
    4. Include irrelevant or nonsensical explanations
    5. Avoid any significant correct information from: "{passage}"
    
    Generate your wrong answer:"""
            else:
                if self.interview_type == 'good':
                    score_range = "between 0.3 and 0.5"
                else:  # mediocre
                    score_range = "between 0.2 and 0.4"
                    
                prompt = f"""Based on this reference passage: "{passage}", generate a partially incorrect answer to this question: "{question}" that would score {score_range} out of 1.0.
    
    Requirements:
    1. Include some minor correct points but major misconceptions
    2. Mix correct and incorrect terminology
    3. Show limited understanding with significant errors
    4. The answer should be relevant but mostly incorrect
    
    Generate your partially incorrect answer now:"""
    
        response = self.llm.invoke(prompt)
        return response.content.strip()


# Add this method to your InterviewPipeline class

    def evaluate_answer(self, question: str, given_answer: str, passage: str) -> Tuple[float, str, str]:
        """Evaluate the answer with enforced scoring ranges based on interview type"""
        evaluation_prompt = f"""Evaluate this answer based on the reference passage, following strict scoring guidelines.
    
    Question: {question}
    Student's Answer: {given_answer}
    Reference Passage: {passage}
    Interview Type: {self.interview_type}
    
    Strict Scoring Guidelines:
    For {self.interview_type} interview type:
    - Good interview: 
      * Correct answers should score 0.8-1.0
      * Incorrect answers should score 0.3-0.5
    - Mediocre interview:
      * Correct answers should score 0.6-0.8
      * Incorrect answers should score 0.2-0.4
    - Poor interview:
      * Correct answers should score 0.4-0.6
      * Incorrect answers should score 0.0-0.2
    
    Provide your evaluation in exactly this format:
    SCORE: (number between 0-1, following the strict ranges above)
    FEEDBACK: (your detailed feedback - be specific about what was correct and incorrect)
    CORRECT_ANSWER: (the complete correct answer)
    
    When scoring:
    1. Be strict about following the scoring ranges
    2. Penalize any incorrect information significantly
    3. For partially correct answers, lean towards the lower end of the range
    4. For incorrect answers, focus on how wrong or irrelevant the answer is
    
    When writing feedback:
    1. Point out specific errors and misconceptions
    2. Identify missing key points
    3. Note any irrelevant information
    4. Explain why the score falls in its range"""
    
        try:
            response = self.llm.invoke(evaluation_prompt)
            content = response.content
    
            # Extract components using regex
            score_match = re.search(r'SCORE:\s*([\d.]+)', content)
            feedback_match = re.search(r'FEEDBACK:\s*(.+?)\s*CORRECT_ANSWER:', content, re.DOTALL)
            correct_answer_match = re.search(r'CORRECT_ANSWER:\s*(.+)', content, re.DOTALL)
    
            if not all([score_match, feedback_match, correct_answer_match]):
                raise ValueError("Response missing required components")
    
            score = float(score_match.group(1))
            feedback = feedback_match.group(1).strip()
            correct_answer = correct_answer_match.group(1).strip()
    
            # Enforce scoring ranges based on interview type
            score_ranges = {
                'good': {'correct': (0.8, 1.0), 'incorrect': (0.3, 0.5)},
                'mediocre': {'correct': (0.6, 0.8), 'incorrect': (0.2, 0.4)},
                'poor': {'correct': (0.4, 0.6), 'incorrect': (0.0, 0.2)}
            }
    
            # Determine if answer should be treated as correct based on score
            is_correct = score >= score_ranges[self.interview_type]['correct'][0]
            valid_range = score_ranges[self.interview_type]['correct' if is_correct else 'incorrect']
            
            # Clamp score to valid range
            score = max(valid_range[0], min(valid_range[1], score))
    
            return score, feedback, correct_answer
    
        except Exception as e:
            print(f"Error in evaluation: {str(e)}")
            return (0.0, "Error evaluating answer", "Unable to provide correct answer due to evaluation error")

    def _select_passage_by_difficulty(self, passages: List[Dict]) -> str:
      """Select a passage that best matches the current difficulty level"""
      for passage in passages:
          if passage['difficulty'].lower() == self.current_difficulty:
              return passage['passage']
      return passages[0]['passage']  # Fallback to first passage if no match

    def _select_review_topic(self, topics_needing_review: set) -> Tuple[str, str]:
        """Select a topic that needs review"""
        return random.choice(list(topics_needing_review))

    def _display_session_summary(self, results: Dict):
        """Display comprehensive session summary"""
        print("\nInterview Session Summary")
        print("="*80)
        print(f"Total Questions: {self.question_count}")
        print(f"Average Score: {results['session_metrics']['average_score']:.2f}")
        print(f"Session Duration: {results['session_metrics']['session_duration']/60:.1f} minutes")
        print(f"Average Time per Question: {results['session_metrics']['avg_time_per_question']:.1f} seconds")
        print(f"Topics Covered: {len(self.covered_topics)}")
        print(f"Topics Needing Review: {len(results['session_metrics']['topics_needing_review'])}")
        print(f"Difficulty Changes: {len(self.difficulty_changes)}")
        print(f"Final Difficulty Level: {self.current_difficulty}")
        print("="*80)

    def generate_performance_plots(self):
        """Generate interactive performance plots"""
        plots_dir = os.path.join(self.save_dir, self.session_id, 'plots')
        os.makedirs(plots_dir, exist_ok=True)

        # Convert data to lists for plotting
        scores = [r['score'] for r in self.performance_log]
        questions = list(range(1, len(scores) + 1))  # Convert range to list
        times = [r['time_taken'] for r in self.performance_log]

        # Create subplots
        fig = make_subplots(rows=2, cols=1,
                           subplot_titles=('Score Progression', 'Time per Question'))

        # Add score progression trace
        fig.add_trace(
            go.Scatter(x=questions, y=scores, mode='lines+markers',
                      name='Score'),
            row=1, col=1
        )

        # Add time per question trace
        fig.add_trace(
            go.Scatter(x=questions, y=times, mode='lines+markers',
                      name='Time (s)'),
            row=2, col=1
        )

        fig.update_layout(
            title=f'{self.interview_type.capitalize()} Interview Performance',
            height=800
        )
        fig.write_html(os.path.join(plots_dir, 'performance_plots.html'))

        # Topic coverage plot - ensure data is properly formatted
        if self.performance_log:
            topic_data = pd.DataFrame(self.performance_log)
            topic_counts = topic_data['topic'].value_counts()
            
            fig = go.Figure(data=[
                go.Bar(x=list(topic_counts.index), y=list(topic_counts.values))
            ])
            fig.update_layout(title='Topics Coverage')
            fig.write_html(os.path.join(plots_dir, 'topic_coverage.html'))

        # Difficulty changes plot
        if self.difficulty_changes:
            diff_changes = pd.DataFrame(self.difficulty_changes)
            fig = go.Figure()
            fig.add_trace(go.Scatter(
                x=list(diff_changes['question']),  # Convert to list
                y=list(diff_changes['new_difficulty']),  # Convert to list
                mode='lines+markers',
                name='Difficulty Level'
            ))
            fig.update_layout(title='Difficulty Progression')
            fig.write_html(os.path.join(plots_dir, 'difficulty_changes.html'))

    def save_results(self):
        """Save interview results and metrics"""
        results_dir = os.path.join(self.save_dir, self.session_id)

        # Save performance log
        with open(os.path.join(results_dir, 'performance_log.json'), 'w') as f:
            json.dump(self.performance_log, f, indent=2)

        # Save metrics
        metrics = {
            'interview_type': self.interview_type,
            'total_score': self.total_score,
            'question_count': self.question_count,
            'average_score': self.total_score / self.question_count if self.question_count > 0 else 0,
            'topics_covered': list(self.covered_topics),
            'difficulty_changes': self.difficulty_changes,
            'session_duration': (datetime.now() - self.session_start_time).total_seconds(),
            'avg_time_per_question': np.mean(self.question_times) if self.question_times else 0
        }

        with open(os.path.join(results_dir, 'metrics.json'), 'w') as f:
            json.dump(metrics, f, indent=2)

    def conduct_interview(self, adaptive_length: bool = True, max_questions: int = 50) -> Dict:
        """Enhanced interview conduct with adaptive length and maximum question limit"""
        self.session_start_time = datetime.now()
        interview_results = []
        topics_needing_review = set()
        all_topics = set(zip(self.passages_df['Topic'], self.passages_df['Sub-Topic']))
    
        while True:
            # Check maximum question limit
            if self.question_count >= max_questions:
                print(f"\nReached maximum question limit ({max_questions}). Ending interview.")
                break
                
            question_start_time = datetime.now()
    
            # Check if all topics are covered
            if len(self.covered_topics) >= len(all_topics):
                # Calculate average score
                avg_score = self.total_score / self.question_count
    
                # Check termination conditions based on interview type
                if (self.interview_type == 'good' and avg_score >= 0.8 and self.question_count >= 5) or \
                   (self.interview_type == 'mediocre' and avg_score >= 0.6 and self.question_count >= 7) or \
                   (self.interview_type == 'poor' and self.question_count >= 10):
                    break
    
                # Reset covered topics if need to continue
                self.covered_topics.clear()
    
            print(f"\nProcessing question {self.question_count + 1}...")
    
            # Select topic and generate question
            topic, subtopic = self.select_next_topic()
            self.current_topic = topic
    
            similar_passages = self.find_similar_passages(f"{topic} {subtopic}")
            selected_passage = self._select_passage_by_difficulty(similar_passages)
    
            question = self.generate_question(
                selected_passage,
                topic,
                subtopic,
                self.current_difficulty
            )
    
            # Generate and evaluate answer
            answer = self.generate_answer(question, selected_passage, True)
            score, feedback, correct_answer = self.evaluate_answer(
                question, answer, selected_passage
            )
    
            # Update metrics
            question_time = (datetime.now() - question_start_time).total_seconds()
            self.question_times.append(question_time)
    
            self.total_score += score
            self.question_count += 1
            self.covered_topics.add((topic, subtopic))
    
            if score < self.correct_threshold[self.interview_type]:
                topics_needing_review.add((topic, subtopic))
    
            # Update difficulty
            old_difficulty = self.current_difficulty
            self.current_difficulty = self.adjust_difficulty(score)
            if old_difficulty != self.current_difficulty:
                self.difficulty_changes.append({
                    'question': self.question_count,
                    'old_difficulty': old_difficulty,
                    'new_difficulty': self.current_difficulty,
                    'score': score
                })
    
            # Store result
            result = {
                'question_number': self.question_count,
                'question': question,
                'answer': answer,
                'correct_answer': correct_answer,
                'score': score,
                'feedback': feedback,
                'topic': topic,
                'subtopic': subtopic,
                'difficulty': self.current_difficulty,
                'time_taken': question_time,
                'was_review': (topic, subtopic) in topics_needing_review
            }
    
            self.performance_log.append(result)
            interview_results.append(result)
    
            # Display result
            self.display_question_result(result)
    
            # Save intermediate results
            self.save_results()
            self.generate_performance_plots()
    
        return {
            'results': interview_results,
            'session_metrics': {
                'total_score': self.total_score,
                'average_score': self.total_score / self.question_count,
                'topics_covered': list(self.covered_topics),
                'topics_needing_review': list(topics_needing_review),
                'difficulty_changes': self.difficulty_changes,
                'session_duration': (datetime.now() - self.session_start_time).total_seconds(),
                'avg_time_per_question': np.mean(self.question_times)
            }
        }
# Example usage and execution
if __name__ == "__main__":
    # Base configuration
    config = {
        
        
        'google_api_key': 'AIzaSyAAieL1Iy2mStPGBKiZ8uMuMofadMuJuO0',
        'passage_db_path': '/kaggle/input/passages/output_retrieved_passages - Sheet1.csv',
        'save_dir': '/kaggle/working/interview_results',
        'model_name': "Jiten1024/Mistral_7b_test_ep_5_r_8_a_64",
        'max_seq_length': 600
    }

    # Create save directory
    os.makedirs(config['save_dir'], exist_ok=True)

    # Run interviews for each type
    combined_results = {}
    for interview_type in ['good', 'mediocre', 'poor']:
        print(f"\nRunning {interview_type} interview type...")

        # Create pipeline instance with current interview type
        pipeline_config = {**config, 'interview_type': interview_type}
        pipeline = InterviewPipeline(**pipeline_config)

        # Conduct interview
        results = pipeline.conduct_interview(adaptive_length=True)
        combined_results[interview_type] = results

        # Save results and generate plots
        pipeline.save_results()
        pipeline.generate_performance_plots()

    # Generate comparative analysis
    comparison_dir = os.path.join(config['save_dir'], 'comparison')
    os.makedirs(comparison_dir, exist_ok=True)

    # Create comparative plots
    fig = make_subplots(rows=2, cols=2,
                       subplot_titles=('Average Scores', 'Question Counts',
                                     'Average Time per Question', 'Topic Coverage'))

    # Average scores comparison
    avg_scores = [results['session_metrics']['average_score']
                 for results in combined_results.values()]
    fig.add_trace(
        go.Bar(x=['good', 'mediocre', 'poor'], y=avg_scores, name='Avg Score'),
        row=1, col=1
    )

    # Question counts comparison
    question_counts = [len(results['results'])
                      for results in combined_results.values()]
    fig.add_trace(
        go.Bar(x=['good', 'mediocre', 'poor'], y=question_counts,
               name='Question Count'),
        row=1, col=2
    )

    # Average time comparison
    # Average time comparison (continued)
    avg_times = [results['session_metrics']['avg_time_per_question']
                for results in combined_results.values()]
    fig.add_trace(
        go.Bar(x=['good', 'mediocre', 'poor'], y=avg_times,
               name='Avg Time'),
        row=2, col=1
    )

    # Topic coverage comparison
    for interview_type, results in combined_results.items():
        topics = [r['topic'] for r in results['results']]
        topic_counts = pd.Series(topics).value_counts()
        fig.add_trace(
            go.Bar(x=topic_counts.index, y=topic_counts.values,
                   name=interview_type),
            row=2, col=2
        )

    fig.update_layout(
        height=1000,
        width=1200,
        title_text="Comparative Analysis of Interview Types",
        showlegend=True
    )
    fig.write_html(os.path.join(comparison_dir, 'comparative_analysis.html'))

    # Save combined metrics
    combined_metrics = {
        interview_type: {
            'average_score': results['session_metrics']['average_score'],
            'total_questions': len(results['results']),
            'avg_time_per_question': results['session_metrics']['avg_time_per_question'],
            'topics_covered': len(set(r['topic'] for r in results['results'])),
            'final_difficulty': results['results'][-1]['difficulty']
        }
        for interview_type, results in combined_results.items()
    }

    with open(os.path.join(comparison_dir, 'combined_metrics.json'), 'w') as f:
        json.dump(combined_metrics, f, indent=2)

    # Generate summary report
    report = f"""Interview Session Summary Report
Generated on: {datetime.now().strftime('%Y-%m-%d %H:%M:%S')}

Performance Summary:
-------------------
"""
    for interview_type, metrics in combined_metrics.items():
        report += f"\n{interview_type.upper()} Interview Type:\n"
        report += f"- Average Score: {metrics['average_score']:.2f}\n"
        report += f"- Total Questions: {metrics['total_questions']}\n"
        report += f"- Average Time per Question: {metrics['avg_time_per_question']:.1f} seconds\n"
        report += f"- Topics Covered: {metrics['topics_covered']}\n"
        report += f"- Final Difficulty Level: {metrics['final_difficulty']}\n"

    report += "\nComparative Analysis:\n"
    report += "--------------------\n"
    report += f"- Best Performance: {max(combined_metrics.items(), key=lambda x: x[1]['average_score'])[0]}\n"
    report += f"- Most Efficient: {min(combined_metrics.items(), key=lambda x: x[1]['avg_time_per_question'])[0]}\n"
    report += f"- Most Topics Covered: {max(combined_metrics.items(), key=lambda x: x[1]['topics_covered'])[0]}\n"

    with open(os.path.join(comparison_dir, 'summary_report.txt'), 'w') as f:
        f.write(report)

    print("\nInterview sessions completed!")
    print(f"Results saved in: {config['save_dir']}")
    print("Generated files include:")
    print("- Individual interview results and plots for each type")
    print("- Comparative analysis plots")
    print("- Combined metrics summary")
    print("- Detailed summary report")
````
